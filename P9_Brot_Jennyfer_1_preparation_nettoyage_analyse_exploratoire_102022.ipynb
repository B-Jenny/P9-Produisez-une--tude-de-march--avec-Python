### 1.2 - Fonction

```python
# Déclaration d'une fonction 

def analyseDataFrame(df):
    """Fonction qui prend en arguments un dataframe pour analyse."""

    name =[x for x in globals() if globals()[x] is df][0]
    print("La dimension de DataFrame « " + name  +" » : " + str(df.shape) + " (lignes, colonnes)\n")
    display(temp_var = df.info())
    display(df.describe().T)
    display(df.head(5))
    
    # Afficher les colonnes qui sont unique dans le DataFrame df
    print("Test d'unicité sur les colonnes de DataFrame :")
    test_unicite = 0
    for col in df.columns :
        if df[col].size == df.drop_duplicates(col)[col].size :
          display(f'« {col:s} » est unique, et pourrait être une clé primaire.')
          test_unicite = 1
    if test_unicite == 0 :
        print("Il n'y a pas d'unicité dans aucune colonne. Il n'y a pas de clé primaire")

    print("Compter les nulles pour chaque colonne :")
    display(df.isnull().sum())
    
    print("Compter les duplicates:")
    display(df.duplicated().sum())
    
```

# *2 - Importation des fichiers et initialisation des DataFrames*

```python
# Importation des données au format csv

Df_dispoAlim_2017 = pd.read_csv("DisponibiliteAlimentaire_2017.csv")
Df_pop_2000_2018 = pd.read_csv("Population_2000_2018.csv")
Df_stabPolitique = pd.read_csv("FAO_StabPolitique.csv")



Df_pib = pd.read_csv("FAO_PIB_V01.csv")
Df_co2 = pd.read_csv("FAO_CO2.csv")
```

## Df_dispoAlim_2017

```python
# Visualisation de Df_dispoAlim_2017 avec la fonction déclarée plus haut

analyseDataFrame(Df_dispoAlim_2017)
```

```python
# Pivot avec agrègation des valeurs en prenant la somme de la variable "Valeur", indexé par "Zone" et "Année"

Df_DispoAlim = pd.pivot_table(Df_dispoAlim_2017, values="Valeur", index=["Zone", "Année"], columns=["Élément"], aggfunc=np.sum).reset_index().copy()
print(Df_DispoAlim.shape) # Donne le nombre de ligne et de colonne du dataFrame
Df_DispoAlim.head(2) # Visualisation des 2 premières lignes du dataFrame
```

```python
# Création d'un dataFrame "Df_temps" contenant uniquement les valeurs concernant "Viande de Volailles" pour la colonne "Élément"

Df_temp = Df_dispoAlim_2017[Df_dispoAlim_2017["Produit"] == "Viande de Volailles"]


#Création d'une colonne "ÉlémentEtProduit" ajouant "V-" devant les valeurs de la colonne "Élément" issue du dataFrame "Df_temps"

Df_temp["ÉlémentEtProduit"] = "V-" + Df_temp["Élément"]


# Création d'un dataFrame "Df_volailles", avec pivot des données de la colonne "ÉlémentEtProduit" et agrègation des valeurs en prenant la somme de la variable "Valeur", indexé par "Zays" et "Année"

Df_volailles = pd.pivot_table(Df_temp, values="Valeur", index=["Zone","Année"], columns=["ÉlémentEtProduit"], aggfunc=np.sum).reset_index().copy()
print(Df_volailles.shape) # Donne le nombre de ligne et de colonne du dataFrame
Df_volailles.head(2)# Visualisation des 2 premières lignes du dataFrame
```

```python
# Création d'un dataFrame "Df_AlimT_Vol", issu d'un Merge externe des dataFrames "Df_DispoAlim" et "Df_volailles"

Df_AlimT_Vol = Df_DispoAlim.merge(Df_volailles, on=["Zone", "Année"], how="outer", indicator=True)
print(Df_AlimT_Vol.shape) # Donne le nombre de ligne et de colonne du dataFrame
Df_AlimT_Vol.head() # Visualisation des 5 premières lignes du dataFrame
```

```python
# Affichage des données uniques de la colonne "_merge" du dataFrame "Df_AlimT_Vol"

print(Df_AlimT_Vol["_merge"].unique())
```

```python
# Création d'un dataFrame "Df_NotExist" avec une nouvelle colonne "Not_Existe" et les colonnes "Zone" et "_merge"

Df_NotExist = pd.DataFrame(columns=["Not_Existe", "Zone", "_merge"])
```

```python
# Création d'une boucle permettant la visualisation des pays n'étant pas présents dans les deux dataFrame

print("Les pays ci-après n'existe pas dans le fichier",'Df_DispoAlim ou Df_volailles (DisponibiliteAlimentaire_2017.csv)')
for index, row in Df_AlimT_Vol[Df_AlimT_Vol['_merge']!="both"][["Zone", "_merge"]].iterrows() :
    if len(Df_volailles[Df_volailles["Zone"] == row["Zone"]]) == 0 :
        if(Df_NotExist[Df_NotExist["Zone"] == row["Zone"]].shape[0] == 0) :
            Df_NotExist = Df_NotExist.append({"Not_Existe": "Df_volailles", "Zone": row["Zone"], "_merge":row["_merge"]}, ignore_index=True)

    if len(Df_DispoAlim[Df_DispoAlim["Zone"] == row["Zone"]]) == 0 :
        if(Df_NotExist[Df_NotExist["Zone"] == row["Zone"]].shape[0] == 0) :
            Df_NotExist = Df_NotExist.append({"Not_Existe": "Df_DispoAlim", "Zone": row["Zone"], "_merge":row["_merge"]}, ignore_index=True)

print(Df_NotExist)
```

```python
# visualisation des valeurs différentes de "both", de la colonne "_merge", du dataFrame "Df_AlimT_Vol" 

Df_AlimT_Vol[Df_AlimT_Vol["_merge"]!="both"]
```

```python

Df_AlimT_Vol = Df_AlimT_Vol.drop(["_merge"], axis=1) # Suppression de la colonne "_merge" du dataFrame "Df_AlimT_Vol"
Df_AlimT_Vol.set_index("Zone", inplace=True) # La colonne "Zone" devient l'index de ligne
Df_AlimT_Vol = Df_AlimT_Vol.fillna(0) # Remplace les valeurs "nan" par "0"
Df_AlimT_Vol.head() # Visualisation des 5 premières lignes du dataFrame
```

```python
# Visualisation du nombre de lignes, de colonnes et du noms des colonnes du dataFrame "Df_AlimT_Vol" 

print(Df_AlimT_Vol.shape)
Df_AlimT_Vol.columns
```

```python
# Calcul du pourcentage de l'Autosuffisance et de l'importation 

Df_AlimT_Vol["V-Ratio_Autosuffisance"] = round((Df_AlimT_Vol["V-Production"] / Df_AlimT_Vol["V-Disponibilité intérieure"])*100,2)
Df_AlimT_Vol["V-Ratio_Importations"] = round((Df_AlimT_Vol["Importations - Quantité"] / Df_AlimT_Vol["V-Disponibilité intérieure"])*100,2)
Df_AlimT_Vol["Ratio_Autosuffisance"] = round((Df_AlimT_Vol["Production"] / Df_AlimT_Vol["Disponibilité intérieure"])*100,2)
Df_AlimT_Vol["Ratio_Importations"] = round((Df_AlimT_Vol["Importations - Quantité"] / Df_AlimT_Vol["Disponibilité intérieure"])*100,2)
```

# Population

```python
# Pivot avec agrègation des valeurs en prenant la somme de la variable "Valeur", indexé par "Zone" et "Année"

Df_popAnnee = pd.pivot_table(Df_pop_2000_2018, values="Valeur", index=["Zone"], columns=["Année"], aggfunc=np.sum).reset_index().copy()
print(Df_popAnnee.shape) # Donne le nombre de lignes et de colonnes du dataFrame "Df_pop_2000_2018"
Df_popAnnee.head() # Visualisation des 5 premières lignes du dataFrame
```

```python
# Création d'une colonne "Croissance_pop" et calcul de la croissance de la population exprimé en pourcentage

Df_popAnnee["Croissance_pop"] = round(((Df_popAnnee[2017] - Df_popAnnee[2013]) / Df_popAnnee[2013])*100,2)
Df_popAnnee # Visualisation du dataFrame

```

```python
# Renommage de la colonne "2017" par "Population_2017"

Df_popAnnee.rename(columns={2017: "Population_2017"}, inplace=True)
```

```python
# Création d'un dataFrame "Df_AlimT_Vol_pop", issu d'un Merge externe des dataFrames "Df_AlimT_Vol" et "Df_popAnnee", concernant uniquement les données liées à l'année "2017"

Df_AlimT_Vol_pop = Df_AlimT_Vol.merge(Df_popAnnee[["Zone", "Croissance_pop", "Population_2017"]], on=["Zone"], how="outer", indicator=True)
print(Df_AlimT_Vol_pop.shape) # Donne le nombre de ligne et de colonne du dataFrame
Df_AlimT_Vol_pop.head(2) # visualisation des 2 premières lignes du dataFrame
```

```python
# Création d'une boucle permettant la visualisation des pays n'étant pas présents dans les deux dataFrame

print("Les pays ci-après n'existe pas dans le fichier","Df_popAnnee ou Df_AlimT_Vol (Population_2000_2018.csv)")
for index, row in Df_AlimT_Vol_pop[Df_AlimT_Vol_pop['_merge']!='both'][['Zone', '_merge']].iterrows() :
    if len(Df_popAnnee[Df_popAnnee["Zone"] == row["Zone"]]) == 0 :
        if(Df_NotExist[Df_NotExist["Zone"] == row["Zone"]].shape[0] == 0) :
            Df_NotExist = Df_NotExist.append({"Not_Existe": "Df_popAnnee", "Zone": row["Zone"], "_merge":row["_merge"]}, ignore_index=True)
    df = Df_AlimT_Vol.reset_index()
    if len(df[df["Zone"] == row["Zone"]]) == 0 :
        if(Df_NotExist[Df_NotExist["Zone"] == row["Zone"]].shape[0] == 0) :
            Df_NotExist = Df_NotExist.append({"Not_Existe": "Df_AlimT_Vol", "Zone": row["Zone"], "_merge":row["_merge"]}, ignore_index=True)
print(Df_NotExist.shape)
print(Df_NotExist)
```

```python
# visualisation des valeurs correspondantes à "both", de la colonne "_merge", du dataFrame "Df_AlimT_Vol_pop" 

Df_AlimT_Vol_pop = Df_AlimT_Vol_pop[Df_AlimT_Vol_pop["_merge"]=="both"]
Df_AlimT_Vol_pop = Df_AlimT_Vol_pop.drop(["_merge"], axis=1) # Supression de la colonne "_merge"
```
